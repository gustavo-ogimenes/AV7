```python
# Importação das bibliotecas necessárias
import numpy as np                # Biblioteca para operações numéricas e vetoriais
import matplotlib.pyplot as plt   # Biblioteca para criação de gráficos

# Definição da função f(x) = sen(x)/x
# Aqui usamos np.where para definir f(0) = 1, evitando divisão por zero
def f(x):
    return np.where(x == 0, 1.0, np.sin(x) / x)

# Criação de um vetor de valores x no intervalo de -10 a 10 com 1000 pontos
# Isso garante um gráfico suave e bem detalhado
x = np.linspace(-10, 10, 1000)

# Avaliação da função f(x) para todos os valores de x gerados
y = f(x)

# Configuração do gráfico
plt.figure(figsize=(10, 5))  # Define o tamanho da figura (10 polegadas por 5)

# Plota a função f(x) = sen(x)/x com linha azul e rótulo para legenda
plt.plot(x, y, label='f(x) = sen(x)/x', color='blue')

# Linhas horizontais e verticais no zero para ajudar na visualização
plt.axhline(0, color='black', linewidth=0.5)  # Linha horizontal em y = 0
plt.axvline(0, color='black', linewidth=0.5)  # Linha vertical em x = 0

# Título e rótulos dos eixos
plt.title('Gráfico da função f(x) = sen(x)/x')
plt.xlabel('x')      # Rótulo do eixo x
plt.ylabel('f(x)')   # Rótulo do eixo y

# Ativa o grid (grade) no fundo do gráfico
plt.grid(True)

# Exibe a legenda no gráfico
plt.legend()

# Exibe o gráfico na tela
plt.show()
```


1. Comentário sobre o comportamento da função:

A função f(x) = sen(x)/x apresenta um comportamento oscilatório, com oscilações que diminuem de amplitude à medida que x se afasta da origem. Isso acontece porque o numerador (a função seno) continua oscilando entre -1 e 1, mas o denominador cresce em módulo, fazendo com que a razão entre os dois vá diminuindo.

À medida que x tende ao infinito (positivo ou negativo), o valor da função se aproxima de zero, o que caracteriza um comportamento de decaimento. A função não é periódica, mas mantém uma oscilação regular com amplitude cada vez menor.

Os zeros da função ocorrem em todos os pontos onde sen(x) é igual a zero, exceto no ponto x = 0, onde a função é definida como 1 por limite. Assim, os zeros ocorrem nos múltiplos não nulos de pi: aproximadamente em x = ±3,14, ±6,28, ±9,42, e assim por diante.




2. Aplicações físicas da função:

A função sen(x)/x, também conhecida como função sinc, aparece em diversos contextos da física e engenharia.

Na área de análise de sinais e telecomunicações, ela representa a resposta ideal de um filtro passa-baixas. Além disso, está diretamente relacionada ao teorema da amostragem de Nyquist, que garante a reconstrução perfeita de um sinal contínuo a partir de amostras discretas.

Em óptica, essa função descreve o padrão de intensidade luminosa gerado pela difração de luz em uma fenda simples. A forma do gráfico está relacionada à distribuição de energia nos experimentos de interferência e difração.

A função também é importante em análise espectral, pois está ligada à transformada de Fourier de uma função retangular, sendo assim uma ponte entre domínios do tempo e da frequência.

```python
# Importação da biblioteca NumPy para cálculos numéricos
import numpy as np

# Definição da função f(x) = sen(x)/x com tratamento correto para x = 0
def f(x):
    y = np.empty_like(x)  # Cria um array com o mesmo formato de x
    y[x == 0] = 1.0        # Define f(0) = 1 diretamente
    y[x != 0] = np.sin(x[x != 0]) / x[x != 0]  # Calcula f(x) normalmente onde x ≠ 0
    return y

# Definição do intervalo de integração
a = -10  # Limite inferior
b = 10   # Limite superior

# Número de subintervalos (quanto maior, mais precisa a aproximação
n = 50

# Geração de n+1 pontos igualmente espaçados entre a e b
x_trap = np.linspace(a, b, n + 1)

# Cálculo dos valores da função f(x) nos pontos gerados
y_trap = f(x_trap)

# Cálculo da largura de cada subintervalo
h = (b - a) / n

# Aplicação da fórmula da Regra do Trapézio:
# Área ≈ (h/2) * [f(x0) + 2*f(x1) + 2*f(x2) + ... + 2*f(x_{n-1}) + f(xn)]
area = (h / 2) * (y_trap[0] + 2 * np.sum(y_trap[1:-1]) + y_trap[-1])

# Exibição do resultado final
print("Área aproximada sob a curva no intervalo [-10, 10]:", area)
```

# 1.Estimativa da área com a regra do trapézio:

A área sob a curva da função f(x) = sen(x)/x, no intervalo de -10 a 10, foi estimada utilizando a regra do trapézio com 50 subintervalos. Esse método consiste em dividir o intervalo em partes iguais e somar áreas aproximadas de trapézios formados entre os pontos da função.

O valor aproximado obtido para a área foi cerca de 3.31459.

# 2.Comentário sobre o resultado:

Análise do Resultado Obtido para a Integral da Função sen(x)/x no Intervalo [-10, 10]

A função sen(x)/x, conhecida como função sinc, é contínua em todo o domínio real, com valor definido como 1 em x = 0, e apresenta oscilações cuja amplitude decresce à medida que o valor absoluto de x aumenta.

O objetivo é calcular a integral definida dessa função no intervalo de -10 a 10, que corresponde à área sob a curva nesse domínio.

Estudos clássicos mostram que a integral imprópria de sen(x)/x no eixo real completo, ou seja, de menos infinito a mais infinito, é igual a π (aproximadamente 3,1416). Sendo assim, espera-se que a integral no intervalo finito [-10, 10], que cobre boa parte das oscilações da função, seja uma aproximação próxima, porém levemente inferior a esse valor.

No entanto, o resultado numérico obtido pelo método da Regra do Trapézio com 50 subintervalos foi aproximadamente 3,31, um valor ligeiramente superior a π e à estimativa esperada em torno de 3,13.

Esse pequeno desvio pode ser explicado por algumas razões:

Natureza da função e método numérico:

A função oscila entre valores positivos e negativos, e a Regra do Trapézio pode, dependendo da discretização dos pontos, capturar um excesso de área positiva em relação à negativa, levando a uma leve superestimação da integral.

Número de subintervalos e refinamento da malha:

Com 50 divisões, o método oferece uma aproximação razoável, mas ainda suscetível a erros de discretização em funções oscilatórias. Aumentar o número de subintervalos ou usar métodos numéricos de maior ordem (como a Regra de Simpson) pode refinar essa aproximação.

Propriedades da integral parcial:

A integral no intervalo [-10, 10] não possui um valor elementar fechado simples, sendo comum que valores numéricos precisos variem ligeiramente conforme o método e parâmetros usados.

Portanto, o valor obtido (aproximadamente 3,31) está dentro de um intervalo aceitável para a técnica numérica empregada, e confirma que a área sob a curva da função sen(x)/x nesse intervalo é da ordem de grandeza de π.

Para melhorar a precisão e validar o resultado, recomenda-se empregar métodos numéricos mais sofisticados ou aumentar a quantidade de subintervalos.
