```python
"""
Análise numérica (com comentários detalhados) das integrais impróprias da imagem:
  1) ∫_0^∞ x^2/(e^x - 1) dx
  2) ∫_0^∞ x^3 e^{-x^2} dx
  3) ∫_0^∞ 1 / sqrt(x(1+x)) dx
  4) ∫_0^∞ cos(x) / (1+x^2) dx
  5) ∫_0^∞ (sin(a x)/x) e^{-b x} dx   (com a=1, b=0.1)

O objetivo: plotar a integranda, justificar intervalo truncado, calcular com
Quadratura de Gauss (obrigatória) + Simpson, comparar com `scipy.integrate.quad`
e com soluções analíticas quando disponíveis.
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad, simps
from scipy.special import roots_legendre
import math

# --- utilitários e configuração ----------------------------------------------------------------

# pequeno epsilon para evitar divisão por zero em pontos singulares (ex.: x=0)
EPS = 1e-12

# função utilitária: quadratura de Gauss-Legendre em [a,b]
def quadratura_gaussiana(f, a, b, n=80):
    """
    Integra f em [a,b] usando n pontos de Gauss-Legendre.
    - roots_legendre(n) devolve os nós xi no intervalo [-1,1] e pesos wi correspondentes.
    - mapeamos xi -> x na forma x = (a+b)/2 + (b-a)/2 * xi
    - o fator (b-a)/2 aparece multiplicando a soma dos pesos.
    Observação: Gauss-Legendre integra exatamente polinômios de grau <= 2n-1. É
    muito eficiente para funções suaves em intervalo finito.
    """
    xi, wi = roots_legendre(n)   # xi: nós em [-1,1], wi: pesos
    xm = 0.5 * (b - a)           # fator de escala (b-a)/2
    xo = 0.5 * (b + a)           # centro (a+b)/2
    x_mapped = xo + xm * xi      # nós mapeados para [a,b]
    return xm * np.sum(wi * f(x_mapped))

# função para estimar divergência observando integrais parciais crescentes
def integra_parcial_sequence(f, a, b_list):
    """
    Calcula ∫_a^{B} f(x) dx para vários B em b_list usando scipy.quad.
    Serve para investigar se a integral tende a um limite ou cresce sem limite.
    """
    resultados = []
    for B in b_list:
        val, err = quad(f, a, B)
        resultados.append((B, val, err))
    return resultados

# --- definindo as funções integrandas (cada uma como função Python) ------------------------------

def f1(x):
    # x**2 / (e^x - 1) -- cuidado em x=0 (limite é finito)
    # usamos EPS perto de 0 para evitar divisão por zero numérica.
    x = np.asarray(x)
    # para x muito pequeno usamos desenvolvimento de Taylor: x^2/(e^x - 1) ~ x
    # mas implementamos de forma segura usando np.where
    denom = np.expm1(x)  # exp(x)-1 com maior precisão para x pequeno
    # quando denom é zero (x=0) substituímos pela limite x -> 0: f ~ x
    safe = np.where(np.abs(denom) < 1e-16, x, x**2 / denom)
    return safe

def f2(x):
    # x^3 * exp(-x^2) -- regular, decai super rápido (e^{-x^2})
    x = np.asarray(x)
    return x**3 * np.exp(-x**2)

def f3(x):
    # 1 / sqrt(x (1+x)) -- singular em x=0 (comportamento ~ 1/sqrt(x)),
    # e decai ~1/x para x grande (o que sugere divergência logarítmica no infinito).
    x = np.asarray(x)
    # para evitar sqrt de zero usamos EPS
    x_safe = np.maximum(x, EPS)
    return 1.0 / np.sqrt(x_safe * (1.0 + x_safe))

def f4(x):
    # cos(x) / (1 + x^2) -- integrável em [0,∞) (decai ~1/x^2) e combinação oscilatória
    x = np.asarray(x)
    return np.cos(x) / (1.0 + x**2)

def f5(x, a=1.0, b=0.1):
    # (sin(a x)/x) * e^{-b x}
    # no ponto x=0 o limite é a (pois sin(ax)/x -> a). Tratamos numericamente.
    x = np.asarray(x)
    # para x muito pequeno substituímos pelo limite a * exp(0) = a
    small = np.abs(x) < 1e-12
    out = np.empty_like(x, dtype=float)
    # evita divisão por zero e usa vetorization
    out[small] = a  # limite x->0 : sin(a x)/x -> a ; multiplicado por e^{-b*0}=1
    out[~small] = (np.sin(a * x[~small]) / x[~small]) * np.exp(-b * x[~small])
    return out

# --- intervalo truncado (escolha e justificativa) ----------------------------------------------
# Escolhemos intervalos [a,b] truncados para aplicar Gauss e Simpson.
# Justificação (resumida):
# 1) f1 decai exponencialmente para x grande (comportamento ~ x^2 e^{-x}), então b ~ 40 é suficiente.
# 2) f2 decai muito rápido (e^{-x^2}), b ~ 10 é suficiente.
# 3) f3 tem singular em 0 e cauda ~ 1/x -> diverge; escolhemos b grande para mostrar crescimento.
# 4) f4 decai ~1/x^2 e é oscilatória; b ~ 40 costuma ser suficiente para convergência.
# 5) f5 com b>0 causa decaimento exponencial leve; b = 200 ou 300 dá boa aproximação para b=0.1.
intervalos_truncados = {
    1: (0.0 + 1e-9, 40.0),
    2: (0.0 + 1e-9, 10.0),
    3: (1e-9, 1e4),   # usamos b grande para evidenciar divergência (parciais)
    4: (0.0 + 1e-9, 40.0),
    5: (0.0 + 1e-9, 300.0)
}

# --- parâmetros numéricos (grades para Simpson, nós para Gauss) ---------------------------------
n_plot = 1000     # número de pontos para plotar a curva
n_simpson = 2000  # número de pontos para Simpson (deve ser grande e uniformemente espaçado)
n_gauss = 80      # nós de Gauss (aumente se quiser mais precisão)

# --- análise para cada integral -----------------------------------------------------------------
for idx in range(1, 6):
    print("\n" + "="*70)
    print(f"Integral ({idx})")
    a, b = intervalos_truncados[idx]

    # Preparar grid para plot/Simpson (grade uniforme em [a,b])
    x_plot = np.linspace(a, b, n_plot)
    x_simps = np.linspace(a, b, n_simpson)

    # selecionar função e (quando houver) valor analítico conhecido
    if idx == 1:
        f = f1
        analytic = None
        # Analítica: ∫_0^∞ x^2/(e^x - 1) dx = Γ(3) ζ(3) = 2 ζ(3)
        # tentaremos obter ζ(3) com scipy.special.zeta (se disponível)
        try:
            from scipy.special import zeta as _zeta
            analytic = 2.0 * _zeta(3.0)
        except Exception:
            analytic = None
        app = "Aplicação: termo que aparece em estatística quântica (radiação de corpo negro, Planck)."
    elif idx == 2:
        f = f2
        analytic = 0.5  # cálculo exato: substituindo u = x^2 -> integral = 1/2
        app = "Aplicação: integrais relacionadas a distribuições Maxwellianas (formas com e^{-x^2})."
    elif idx == 3:
        f = f3
        analytic = None  # não existe valor finito: integral diverge (comportamento ~ 1/x para x grande)
        app = "Aplicação: funções de densidade que podem aparecer em problemas de densidade de estados (ex.: integrandas com singular)."
    elif idx == 4:
        f = f4
        # Analítica: ∫_0^∞ cos(x)/(1+x^2) dx = (π/2) e^{-1}
        analytic = 0.5 * math.pi * math.exp(-1.0)
        app = "Aplicação: resposta de sistemas oscilatórios (transformadas, convoluções)."
    elif idx == 5:
        # parámentros a e b pedidos: a=1, b=0.1
        a_param = 1.0
        b_param = 0.1
        f = lambda x: f5(x, a=a_param, b=b_param)
        # Analítica: ∫_0^∞ e^{-b x} sin(a x)/x dx = arctan(a/b)  (para b>0)
        analytic = math.atan(a_param / b_param)
        app = "Aplicação: integrandos que aparecem em transformadas de Laplace e em respostas amortecidas."

    # --- Ploting da função integranda (para inspeção visual) -----------------------------------
    # Em caso de singularidades próximas de zero, usamos y_plot com limites bem comportados.
    with np.errstate(all='ignore'):
        y_plot = f(x_plot)

    plt.figure(figsize=(7,3.5))
    plt.plot(x_plot, y_plot, lw=1)
    plt.title(f"Integranda da integral ({idx}) — visualizar comportamento")
    plt.xlabel("x")
    plt.tight_layout()
    plt.grid(True)
    plt.show()

    # --- Cálculo pela Quadratura Gaussiana (no intervalo truncado [a,b]) ------------------------
    # Observação: esse valor é a integral *truncada* em [a,b]; para integrar em (0,∞) usamos quad com inf
    try:
        gauss_val = quadratura_gaussiana(f, a, b, n=n_gauss)
    except Exception as e:
        gauss_val = np.nan
        print("Erro na quadratura gaussiana:", e)

    # --- Cálculo por Simpson (usando grade uniforme em [a,b]) -----------------------------------
    # simps exige valores em grade uniforme; se n_simpson for par, simps funciona bem (não depende de ser par).
    y_simps = f(x_simps)
    simpson_val = simps(y_simps, x_simps)

    # --- Cálculo de referência com scipy.integrate.quad (se aplicável) --------------------------
    # quad suporta intervalos improprios via limite superior np.inf.
    quad_val = None
    quad_err = None
    if idx != 3:
        # para as integrais onde esperamos convergência, usamos quad(a, inf) como referência
        try:
            quad_val, quad_err = quad(f, a, np.inf, limit=200)
        except Exception as e:
            quad_val = None
            quad_err = None
            print("quad falhou:", e)
    else:
        # para o caso 3 (sabemos que diverge), vamos calcular parciais para demonstrar crescimento
        B_list = [10.0, 100.0, 1000.0, 10000.0]
        print("Avaliação de parciais para demonstrar divergência (integral 3):")
        parciais = integra_parcial_sequence(f, a, B_list)
        for B, val, err in parciais:
            print(f"  ∫_{a:.1e}^{B:.1e} f(x) dx ≈ {val:.8g} (erro ≈ {err:.2g})")
        # não definimos quad_val, pois integramos parciais apenas

    # --- Comparação e saída impressa ------------------------------------------------------------
    print(f"Intervalo truncado usado para Gauss/Simpson: [{a:.1e}, {b:.1e}]")
    print(f"Quadratura de Gauss (n={n_gauss}): {gauss_val}")
    print(f"Simpson (grade {n_simpson} pontos): {simpson_val}")
    if quad_val is not None:
        print(f"SciPy quad (referência em (0,∞)): {quad_val} (estimativa de erro: {quad_err})")
    if analytic is not None:
        print(f"Valor analítico conhecido (quando aplicável): {analytic}")
    # diferenças
    if analytic is not None and quad_val is not None:
        print(f"Abs(quad - analítico) = {abs(quad_val - analytic):.3e}")
    if analytic is not None:
        print(f"Abs(Gauss - analítico) = {abs(gauss_val - analytic):.3e}")
        print(f"Abs(Simpson - analítico) = {abs(simpson_val - analytic):.3e}")

    # comentário sobre convergência/divergência
    if idx == 3:
        print("Conclusão (3): a integral diverge — a cauda se comporta ~ 1/x, produzindo crescimento logarítmico.")
    else:
        # se quad devolveu um número finito, a integral converge
        if quad_val is not None and np.isfinite(quad_val):
            print("Conclusão: a integral converge (valor finito obtido por quad).")
        else:
            print("Conclusão: não foi possível confirmar convergência numericamente com quad aqui.")

    # aplicação física resumida
    print("Nota de aplicação:", app)

# --- Observações finais úteis -------------------------------------------------------------------
print("\nObservações finais:")
print("- A quadratura de Gauss é muito eficiente em intervalos finitos e para funções suaves.")
print("- Para integrais impróprias em (0,∞) fazemos truncamento [0,B] e/ou usamos quad(..., np.inf).")
print("- Quando existe singularidade em 0, trabalhe com limite (por exemplo substituindo x=0 por pequeno EPS),")
print("  ou use transformações analíticas se necessário.")
print("- Para o caso (3) mostramos parciais que crescem quando B aumenta, indicando divergência.")
print("- Ajuste n_gauss, n_simpson e os limites de truncamento para equilibrar tempo de execução vs. precisão.")
```

A terceira função analisada foi a que apresentou maior dificuldade de convergência para ambos os métodos. Isso aconteceu porque ela não converge para valores muito grandes de x, já que seu comportamento tende a decrescer de forma lenta, o que resulta em um crescimento indefinido. Nos métodos numéricos, isso se refletiu em resultados que aumentavam continuamente à medida que o intervalo de truncamento era expandido, sem que houvesse estabilização do valor.

O comportamento da função para valores muito grandes ou muito próximos de zero influenciou diretamente a escolha dos intervalos de truncamento. Para funções com decaimento rápido, como nas integrais que envolvem termos exponenciais, foi possível adotar valores relativamente pequenos para o limite superior, pois a contribuição da parte final da curva era praticamente desprezível. Já funções com cauda lenta exigiram limites muito maiores para que a maior parte da contribuição fosse incluída, ainda que, em alguns casos, isso apenas confirmasse a divergência. Quando havia comportamento problemático próximo de zero, como singularidades, foi necessário substituir o ponto por um valor muito pequeno ou recorrer a aproximações específicas para evitar instabilidade numérica.

A quadratura gaussiana mostrou-se mais adequada para funções suaves e bem comportadas em intervalos finitos, especialmente quando o decaimento rápido permitia truncar o intervalo sem perda significativa de precisão. O método foi menos eficiente para funções com singularidades internas ou com comportamento oscilatório prolongado.

O cálculo numérico de integrais impróprias exige alguns cuidados. É fundamental escolher um limite de truncamento que capture a maior parte da contribuição da função sem desperdiçar recursos computacionais. Singularidades precisam ser tratadas com deslocamentos ou aproximações que garantam estabilidade. Também é importante verificar a convergência por meio do cálculo de integrais parciais e analisar se o valor obtido se mantém estável. A escolha do método deve considerar o comportamento da função: métodos de passo fixo podem ser ineficientes em variações rápidas, enquanto métodos adaptativos tendem a se ajustar melhor.

Compreender o comportamento qualitativo da função antes de aplicar métodos numéricos é essencial. Isso ajuda a identificar possíveis divergências, selecionar o método mais apropriado, definir intervalos de truncamento eficientes e interpretar corretamente os resultados, evitando confundir erros numéricos com características reais do problema.

Para cada método, o intervalo de integração foi escolhido considerando o comportamento da função no domínio e a necessidade de minimizar erros de truncamento.

Em funções com comportamento singular próximo de zero, o limite inferior não foi definido exatamente no ponto da singularidade, mas sim deslocado para um valor muito pequeno, evitando instabilidades numéricas. Esse ajuste permitiu que os métodos operassem com maior estabilidade e precisão.

Assim, cada intervalo foi escolhido de modo a equilibrar eficiência computacional e fidelidade ao comportamento real da função, minimizando o impacto de truncamento e preservando a representatividade do resultado.
