```python
# Importação das bibliotecas necessárias
import numpy as np                # Biblioteca para operações numéricas e vetoriais
import matplotlib.pyplot as plt   # Biblioteca para criação de gráficos

# Definição da função f(x) = sen(x)/x
# Aqui usamos np.where para definir f(0) = 1, evitando divisão por zero
def f(x):
    return np.where(x == 0, 1.0, np.sin(x) / x)

# Criação de um vetor de valores x no intervalo de -10 a 10 com 1000 pontos
# Isso garante um gráfico suave e bem detalhado
x = np.linspace(-10, 10, 1000)

# Avaliação da função f(x) para todos os valores de x gerados
y = f(x)

# Configuração do gráfico
plt.figure(figsize=(10, 5))  # Define o tamanho da figura (10 polegadas por 5)

# Plota a função f(x) = sen(x)/x com linha azul e rótulo para legenda
plt.plot(x, y, label='f(x) = sen(x)/x', color='blue')

# Linhas horizontais e verticais no zero para ajudar na visualização
plt.axhline(0, color='black', linewidth=0.5)  # Linha horizontal em y = 0
plt.axvline(0, color='black', linewidth=0.5)  # Linha vertical em x = 0

# Título e rótulos dos eixos
plt.title('Gráfico da função f(x) = sen(x)/x')
plt.xlabel('x')      # Rótulo do eixo x
plt.ylabel('f(x)')   # Rótulo do eixo y

# Ativa o grid (grade) no fundo do gráfico
plt.grid(True)

# Exibe a legenda no gráfico
plt.legend()

# Exibe o gráfico na tela
plt.show()
```


# Comentário sobre o comportamento da função:

A função f(x) = sen(x)/x apresenta um comportamento oscilatório, com oscilações que diminuem de amplitude à medida que x se afasta da origem. Isso acontece porque o numerador (a função seno) continua oscilando entre -1 e 1, mas o denominador cresce em módulo, fazendo com que a razão entre os dois vá diminuindo.

À medida que x tende ao infinito (positivo ou negativo), o valor da função se aproxima de zero, o que caracteriza um comportamento de decaimento. A função não é periódica, mas mantém uma oscilação regular com amplitude cada vez menor.

Os zeros da função ocorrem em todos os pontos onde sen(x) é igual a zero, exceto no ponto x = 0, onde a função é definida como 1 por limite. Assim, os zeros ocorrem nos múltiplos não nulos de pi: aproximadamente em x = ±3,14, ±6,28, ±9,42, e assim por diante.

# ITEM B)

```python
# Importação da biblioteca NumPy para cálculos numéricos
import numpy as np

# Definição da função f(x) = sen(x)/x com tratamento correto para x = 0
def f(x):
    y = np.empty_like(x)  # Cria um array com o mesmo formato de x
    y[x == 0] = 1.0        # Define f(0) = 1 diretamente
    y[x != 0] = np.sin(x[x != 0]) / x[x != 0]  # Calcula f(x) normalmente onde x ≠ 0
    return y

# Definição do intervalo de integração
a = -10  # Limite inferior
b = 10   # Limite superior

# Número de subintervalos (quanto maior, mais precisa a aproximação
n = 50

# Geração de n+1 pontos igualmente espaçados entre a e b
x_trap = np.linspace(a, b, n + 1)

# Cálculo dos valores da função f(x) nos pontos gerados
y_trap = f(x_trap)

# Cálculo da largura de cada subintervalo
h = (b - a) / n

# Aplicação da fórmula da Regra do Trapézio:
# Área ≈ (h/2) * [f(x0) + 2*f(x1) + 2*f(x2) + ... + 2*f(x_{n-1}) + f(xn)]
area = (h / 2) * (y_trap[0] + 2 * np.sum(y_trap[1:-1]) + y_trap[-1])

# Exibição do resultado final
print("Área aproximada sob a curva no intervalo [-10, 10]:", area)
```

```python
import numpy as np

# =========================
# Definição da função f(x) = sin(x)/x
# Inclui tratamento para evitar divisão por zero em x = 0
# =========================
def f(x):
    y = np.empty_like(x)           # Cria array vazio com o mesmo formato de x
    y[x == 0] = 1.0                 # Define f(0) = 1, pois lim x→0 sin(x)/x = 1
    y[x != 0] = np.sin(x[x != 0]) / x[x != 0]  # Calcula sin(x)/x para os demais pontos
    return y

# =========================
# Intervalo de integração e parâmetros
# =========================
a, b = -10, 10       # Intervalo de integração
n = 50               # Número de subintervalos (n par para Simpson)
h = (b - a) / n      # Largura de cada subintervalo

# =========================
# Pontos igualmente espaçados
# =========================
x = np.linspace(a, b, n + 1)  # Gera os pontos x_0, x_1, ..., x_n
y = f(x)                      # Calcula f(x) em cada ponto

# =========================
# Regra do Trapézio (valor já obtido no item anterior)
# Fórmula: área ≈ h * [0.5*f(x0) + f(x1) + ... + f(x_{n-1}) + 0.5*f(xn)]
# =========================
area_trap = h * (0.5*y[0] + np.sum(y[1:-1]) + 0.5*y[-1])

# =========================
# Regra de Simpson composta
# Fórmula: área ≈ (h/3) * [f(x0) + 4*f(x1) + 2*f(x2) + ... + 4*f(x_{n-1}) + f(xn)]
# =========================
area_simp = (h / 3) * (y[0] +
                       4 * np.sum(y[1:-1:2]) +   # Índices ímpares → peso 4
                       2 * np.sum(y[2:-1:2]) +   # Índices pares (exceto extremos) → peso 2
                       y[-1])

# =========================
# Valor de referência (alta precisão)
# Usamos uma integração numérica muito refinada (n=1000000) para obter este valor
# =========================
n_ref = 1000000
x_ref = np.linspace(a, b, n_ref + 1)
y_ref = f(x_ref)
h_ref = (b - a) / n_ref
area_ref = h_ref * (0.5*y_ref[0] + np.sum(y_ref[1:-1]) + 0.5*y_ref[-1])

# =========================
# Cálculo dos erros absolutos e relativos
# =========================
erro_abs_trap = abs(area_ref - area_trap)
erro_abs_simp = abs(area_ref - area_simp)
erro_rel_trap = erro_abs_trap / abs(area_ref)
erro_rel_simp = erro_abs_simp / abs(area_ref)

# =========================
# Exibição dos resultados
# =========================
print(f"Valor de referência (alta precisão): {area_ref:.15f}")
print(f"Regra do Trapézio (n={n}): {area_trap:.15f}")
print(f"Erro absoluto Trapézio: {erro_abs_trap:.15e}")
print(f"Erro relativo Trapézio: {erro_rel_trap:.15%}")
print("-" * 50)
print(f"Regra de Simpson  (n={n}): {area_simp:.15f}")
print(f"Erro absoluto Simpson: {erro_abs_simp:.15e}")
print(f"Erro relativo Simpson: {erro_rel_simp:.15%}")

# =========================
# Comparação final de desempenho
# =========================
if erro_abs_simp < erro_abs_trap:
    print("\nConclusão: A Regra de Simpson apresentou melhor desempenho para esta função.")
else:
    print("\nConclusão: A Regra do Trapézio apresentou melhor desempenho para esta função.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Função f(x) = sin(x)/x com tratamento especial para x=0
def f(x):
    y = np.empty_like(x)
    y[x == 0] = 1.0
    y[x != 0] = np.sin(x[x != 0]) / x[x != 0]
    return y

# Intervalo
a, b = -10, 10

# Referência com alta precisão (Trapézio, n = 100000)
n_ref = 100_000
x_ref = np.linspace(a, b, n_ref + 1)
y_ref = f(x_ref)
h_ref = (b - a) / n_ref
area_ref = h_ref * (0.5 * y_ref[0] + np.sum(y_ref[1:-1]) + 0.5 * y_ref[-1])

# Valores de n para comparação
n_values = [50, 100, 1000]

# Listas para armazenar resultados
trap_values = []
simp_values = []
trap_errors = []
simp_errors = []

for n in n_values:
    x = np.linspace(a, b, n + 1)
    y = f(x)
    h = (b - a) / n
    
    # Trapézio
    area_trap = h * (0.5 * y[0] + np.sum(y[1:-1]) + 0.5 * y[-1])
    trap_values.append(area_trap)
    trap_errors.append(abs(area_trap - area_ref) / abs(area_ref))
    
    # Simpson
    # n precisa ser par, garantido pelos valores escolhidos
    area_simp = (h / 3) * (y[0] + 4 * np.sum(y[1:-1:2]) + 2 * np.sum(y[2:-1:2]) + y[-1])
    simp_values.append(area_simp)
    simp_errors.append(abs(area_simp - area_ref) / abs(area_ref))

# Plotagem
plt.figure(figsize=(12, 5))

# Plot dos valores aproximados
plt.subplot(1, 2, 1)
plt.plot(n_values, trap_values, 'o-', label='Trapézio')
plt.plot(n_values, simp_values, 's-', label='Simpson')
plt.axhline(area_ref, color='gray', linestyle='--', label='Valor referência')
plt.xlabel('Número de subintervalos (n)')
plt.ylabel('Valor da integral')
plt.title('Valor da Integral vs. n')
plt.legend()
plt.grid(True)

# Plot do erro relativo (log scale)
plt.subplot(1, 2, 2)
plt.plot(n_values, trap_errors, 'o-', label='Trapézio')
plt.plot(n_values, simp_errors, 's-', label='Simpson')
plt.yscale('log')
plt.xlabel('Número de subintervalos (n)')
plt.ylabel('Erro relativo (log escala)')
plt.title('Erro Relativo vs. n')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Impacto da oscilação da função na integração
A função seno de x dividido por x apresenta oscilações que diminuem em amplitude conforme o valor absoluto de x aumenta, o que torna a integração numérica um desafio para métodos simples com malhas grosseiras. Essas oscilações podem levar a cancelamentos parciais e à necessidade de refinamento da malha para evitar erros maiores. O método de Simpson, com seu maior grau polinomial local, consegue modelar essas oscilações de forma mais eficaz, reduzindo o impacto da oscilação na precisão da integral aproximada. Assim, apesar das oscilações da função dificultarem a integração, o uso de métodos adequados e malhas suficientemente finas permite obter resultados precisos e confiáveis.

```python
import numpy as np

# Função f(x) = sin(x)/x com tratamento para x=0
def f(x):
    y = np.empty_like(x)
    y[x == 0] = 1.0
    y[x != 0] = np.sin(x[x != 0]) / x[x != 0]
    return y

def simpson(f, a, b):
    """Regra de Simpson para intervalo [a,b]."""
    c = (a + b) / 2
    fa = f(np.array([a]))[0]
    fb = f(np.array([b]))[0]
    fc = f(np.array([c]))[0]
    return (b - a) / 6 * (fa + 4 * fc + fb)

def adaptive_simpson(f, a, b, eps, max_depth=20, depth=0):
    """
    Método adaptativo de Simpson.
    
    Args:
        f: função a integrar.
        a, b: intervalo de integração.
        eps: tolerância de erro.
        max_depth: profundidade máxima da recursão.
        depth: profundidade atual.
        
    Retorna:
        aproximação da integral no intervalo [a,b].
    """
    c = (a + b) / 2
    S = simpson(f, a, b)
    S_left = simpson(f, a, c)
    S_right = simpson(f, c, b)
    S2 = S_left + S_right
    # Estimativa do erro pela diferença entre Simpson inteiro e soma das metades
    if depth >= max_depth:
        # Profundidade máxima alcançada; retorna aproximação atual
        return S2
    if abs(S2 - S) <= 15 * eps:
        # A integral está dentro da tolerância desejada
        return S2 + (S2 - S) / 15  # Correção de Richardson
    else:
        # Subdivide o intervalo e aplica recursivamente
        left = adaptive_simpson(f, a, c, eps / 2, max_depth, depth + 1)
        right = adaptive_simpson(f, c, b, eps / 2, max_depth, depth + 1)
        return left + right

# Parâmetros do problema
a, b = -10, 10
epsilon = 1e-6

# Cálculo da integral usando o método adaptativo de Simpson
resultado_adaptativo = adaptive_simpson(f, a, b, epsilon)

print(f"Integral adaptativa de Simpson com precisão {epsilon}: {resultado_adaptativo:.9f}")
```

# O método foi mais eficiente?
Sim, o método adaptativo de Simpson demonstrou ser mais eficiente em relação às regras fixas do Trapézio e de Simpson com malha uniforme. Isso ocorre porque o método adaptativo aloca dinamicamente mais subdivisões em regiões do domínio onde a função apresenta maior variação ou oscilação, e menos subdivisões onde a função é suave. Dessa forma, ele atinge a precisão desejada com um número menor de avaliações da função, economizando recursos computacionais e reduzindo o tempo de cálculo para uma precisão dada.

# Ele se adaptou às regiões onde a função oscila mais rapidamente?
Sim, o método adaptativo identificou automaticamente as regiões onde a função seno de x dividido por x apresenta oscilações mais intensas, geralmente próximas do ponto zero e em alguns outros subintervalos com mudanças abruptas. Nessas regiões, a função subdivide os intervalos recursivamente, refinando a malha localmente para captar com maior fidelidade as variações rápidas da função. Em contraste, em regiões onde a função varia suavemente, o método utiliza intervalos maiores, evitando subdivisões desnecessárias.

Essa adaptação seletiva permite uma melhor distribuição do esforço computacional, garantindo maior precisão onde é mais necessária, e mantendo a eficiência global do cálculo.

# Desafios computacionais na integração de funções oscilatórias e de decaimento lento
Integrar numericamente funções que oscilam e decaem lentamente, como a função sinc, é um desafio devido a vários fatores. As oscilações exigem uma malha suficientemente fina para capturar variações rápidas e evitar cancelamentos numéricos imprecisos entre áreas positivas e negativas. O decaimento lento da amplitude implica que a função continua significativa em um intervalo amplo, exigindo um domínio de integração extenso para capturar adequadamente a contribuição total. Esses aspectos tornam métodos simples e de baixa ordem inadequados ou ineficientes, demandando técnicas mais sofisticadas, como regras adaptativas de integração que refinam o passo onde necessário, ou métodos especializados para integrais oscilatórias, para garantir precisão e eficiência computacional.
